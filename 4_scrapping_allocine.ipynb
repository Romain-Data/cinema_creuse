```python
import pandas as pd

# Charger le fichier Google Sheet
url = 'https://docs.google.com/spreadsheets/d/1B4gkJX_H6kHkVrQxghuc5hSep3pdeqLOnVWQo6vzd2c/export?format=csv'
df = pd.read_csv(url)

# Calculer la médiane de la densité de population
median_density = df['Pop. Density (per sq. km2)'].median()

# Créer un tableau croisé dynamique
pivot_table = pd.pivot_table(df, index=['Country'], columns=['Above/Below Median'], values=['Pop. Density (per sq. km2)'], aggfunc='count')


# Remplir les valeurs manquantes avec 0
pivot_table = pivot_table.fillna(0)

# Renommer les colonnes
pivot_table.columns = ['Below Median', 'Above Median']

# Trier le tableau par les pays au-dessus et en dessous de la médiane
pivot_table = pivot_table.sort_values(by=['Above Median', 'Below Median'], ascending=False)

print(pivot_table)

```
