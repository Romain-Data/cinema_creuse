```python
import pandas as pd
```

```python
df_film_fr = pd.read_csv('films_fr.csv')
df_film_fr.head()
```

```python
df_notes = pd.read_csv('notes_filtrees.csv')
df_notes.head()
```

```python
df_FR_notes = df_notes.merge(
    right = df_film_fr,
    how='inner',
    left_on='tconst',
    right_on='titleId'
)
```

```python
df_FR_notes.drop('titleId',axis=1,inplace=True)
df_FR_notes.drop(['types','attributes','isOriginalTitle'],axis=1,inplace=True)
df_FR_notes.drop(['Unnamed: 0_x','Unnamed: 0_y','ordering'],axis=1,inplace=True)
df_FR_notes.info()
```

```python
df_FR_notes.to_csv('film_fr_notes.csv')
```
