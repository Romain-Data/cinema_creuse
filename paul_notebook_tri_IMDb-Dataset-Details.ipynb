# Etape 1 : df_title_basics

**Retirer tous les films pour adultes.**

dans la colonne 'isAdult' on ne devrait y avoir que des booléens (0/1).
Vérifions s'il y a autre chose

```python
display(df_title_basics['isAdult'].unique())
len(df_title_basics['isAdult'].unique())
```

Il n'y a pas que des booléens.
Il y a 0/1 en int (boolean) et 0/1 en str

Conversion de 'isAdult' en int et verifions que nous n'avons que deux elements de type int dans cette colonne.

```python
df_title_basics['isAdult'] = pd.to_numeric(df_title_basics['isAdult'], errors='coerce').fillna(0).astype(int)
display(df_title_basics['isAdult'].unique())
len(df_title_basics['isAdult'].unique())
```

Suppression de tous les films dont __'isAdult'__ vaut 1.
On peut aisément imaginer que tous les films __isAdult==1__ ne sont pas forcément à caractere pornographique, mais les données ne sont pas forcément bien remplies, certains films X sont spécifiés dans le genre et dautres non.

```python
nb_films_is_adult = len(df_title_basics[df_title_basics['isAdult'] == 1])
print(f"Nombre de films classés adult = {format(nb_films_is_adult, ',').replace(',',' ')}\nCa represente {round(100*nb_films_is_adult/nombre_de_lignes_title_basics_corriges,2)} % du nombre de films total dans df_title_basics")
```

3% -> decision : suppression de tous les films __isAdult == 1__

```python
print(f"Nombre de lignes dans le DataFrame 'df_title_basics' : {format(nombre_de_lignes_title_basics_corriges, ',').replace(',',' ')}")
df_tampon = pd.DataFrame()
df_tampon = df_title_basics[(df_title_basics['isAdult'] == 0)]
df_title_basics_v1 = df_tampon
nombre_de_lignes_title_basics_v1 = df_title_basics_v1.shape[0]
print(f"Nombre de lignes dans le DataFrame 'df_title_basics_v1' : {format(nombre_de_lignes_title_basics_v1, ',').replace(',',' ')}")
print(f"Difference : {format(nombre_de_lignes_title_basics_corriges - nombre_de_lignes_title_basics_v1, ',').replace(',',' ')}")
```

__Faire des visualisations pertinentes de notes et de nombre de votants afin de sortir des criteres de tri__

```python
# Graphique de distribution des averageRating
plt.figure(figsize=(10, 6))
sns.histplot(df_title_ratings['averageRating'], bins=30, kde=True)
plt.title('Distribution des averageRating')
plt.xlabel('Note Moyenne')
plt.ylabel('Fréquence')

# Highlight sur l'intervalle [5;10]
plt.axvspan(5, 10, color='red', alpha=0.3, label='Intervalle [5;10]')

# Légende
plt.legend()

plt.show()
```

```python
# Médiane de numVotes
median_numVotes = df_title_ratings['numVotes'].median()

# Masque pour les données à highlighter
highlight = df_title_ratings[(df_title_ratings['averageRating'] >= 5) & (df_title_ratings['numVotes'] >= median_numVotes)]

# Scatter plot des données complètes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='numVotes', y='averageRating', data=df_title_ratings, label='Données complètes')

# Scatter plot des données à highlighter
sns.scatterplot(x='numVotes', y='averageRating', data=highlight, color='red', label='Notes >= 5 & numVotes >= Médiane')

# Titres et labels
plt.title('Relation entre numVotes et averageRating')
plt.xlabel('Nombre de Votes')
plt.ylabel('Note Moyenne')
plt.legend()
plt.show()
```

__On garde que les films qui ont une note >= 5/10__

```python
print(f"Nombre de lignes dans le DataFrame 'df_title_ratings' : {format(nombre_de_lignes_title_ratings_corriges, ',').replace(',',' ')}")
df_title_ratings_v1 = df_title_ratings[df_title_ratings['averageRating'] >= 5]
nombre_de_lignes_v1 = df_title_ratings_v1.shape[0]
print(f"Nombre de lignes dans le DataFrame 'df_title_ratings_v1' : {format(nombre_de_lignes_v1, ',').replace(',',' ')}")
print(f"Difference' : {format(nombre_de_lignes_title_ratings_corriges - nombre_de_lignes_v1, ',').replace(',',' ')}")
```

__On garde que les films qui ont un nombre de votes >= moyenne de vote__
On aurait pu prendre la médiane mais elle est assez faible, on fait le choix de ne garder que des films qui ont été évalué par un nombre conséquent de personnes.

```python
# plus petit nombre de votes
plus_petit_nombre_de_votes = df_title_ratings_v1['numVotes'].min()
print(f'plus petit nombre de vote = {plus_petit_nombre_de_votes}')

# plus grand nombres de votes
plus_grand_nombre_de_votes = df_title_ratings_v1['numVotes'].max()
print(f'plus grand nombre de vote = {plus_grand_nombre_de_votes}')

# moyenne du nombre de votes
moyenne_nombre_de_votes = df_title_ratings_v1['numVotes'].mean()
print(f'moyenne du nombre de vote = {moyenne_nombre_de_votes}')

# mediane nombre de votes
mediane_nombre_de_votes = df_title_ratings_v1['numVotes'].median()
print(f'mediane nombre de vote = {mediane_nombre_de_votes}')

# nombre de films dont le nombre de vote est superieur à la moyenne
nb_films_nb_vote_sup_median = (df_title_ratings_v1['numVotes'] >= mediane_nombre_de_votes).sum()
print(f'Nombre de films dont le nombre de vote est superieur à la moyenne = {nb_films_nb_vote_sup_median}')

df_title_ratings_v2 = df_title_ratings_v1[(df_title_ratings_v1['numVotes'] >= mediane_nombre_de_votes)]


print(f"\n\nNombre de lignes dans le DataFrame 'df_title_ratings_v1' : {format(nombre_de_lignes_v1, ',').replace(',',' ')}")
nombre_de_lignes_v2 = df_title_ratings_v2.shape[0]
print(f"Nombre de lignes dans le DataFrame 'df_title_ratings_v2' : {format(nombre_de_lignes_v2, ',').replace(',',' ')}")
print(f"Difference' : {format(nombre_de_lignes_v1 - nombre_de_lignes_v2, ',').replace(',',' ')}")
```
