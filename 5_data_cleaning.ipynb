```python
import pandas as pd
import plotly.express as px
import re
```

```python
df = pd.read_csv('film_serie_fr_notes.csv')
df.drop("Unnamed: 0",axis=1,inplace=True)
df.info()
```

# Définir la décennie

```python
filtre = df['startYear']=='\\N'
df[filtre]
```

```python
# Compléter les valeurs manquantes
valeurs_manquantes ={
    "Crime dans l'Hérault" : 2019,
    "Star Wars: Detours" : 2012
}
for film, annee in valeurs_manquantes.items():
    df.loc[df['title']== film,['startYear']] = annee
    display(df.loc[df['title']== film])
```

```python
df['startYear'] = df['startYear'].astype(int)
```

```python
def decade(annee):
    return int(annee)//10*10
df['decade'] = df['startYear'].apply(decade)
```

```python
filtre = df['isAdult'] == 0
df = df[filtre]
```

```python
filtre = df['title'].str.contains('Harry', case = False)
df[filtre]
```

# "Explode" des genres

```python
liste_genre = df['genres'].to_list()
#flat_list = [el[2:] for el in liste_genre]
for i, el in enumerate(liste_genre):
    liste_genre[i] = re.findall('[a-zA-Z].+[a-z]', el)
flat_list=[]
for el in liste_genre:
    flat_list += el

flat_list = list(set(flat_list))
flat_again=[]
for el in flat_list:
    flat_again.extend((el.split(',')))
flat_again = [el.strip() for el in flat_again]
flat_again = [re.sub('[^a-zA-Z]','',el) for el in flat_again]
genres = list(set(flat_again))
```

```python
for i, el in enumerate(genres):
    if el == 'SciFi':
        genres[i] = 'Sci-fi'
    elif el == 'FilmNoir':
        genres[i] = 'Film Noir'
genres
```

```python
df = df.reindex(columns=list(df.columns)+genres)
df
```

```python
df[df.columns[14:]] = df[df.columns[14:]].fillna(value=0)
df[df.columns[14:]]
```

```python
for genre in genres:
    df.loc[df['genres'].str.contains(genre), genre] = 1
df.head()
```
